##26. Contador: Software contable, cartera de clientes, estados financieros, número de cédula, impuestos gestionados y bufete

```python
# ======================================================
# SISTEMA DE CONTADORES
# CBTIS 128
# Especialidad: Inteligencia Artificial
# Submodulo 2: Elabora proyectos de programación lógica
# Docente: Eliseo Nava
# Alumno: Ian Josafat Gutierrez Saucedo
# ======================================================


# ======================================================
# CLASE CONTADOR
# ======================================================

class contador:

    # -----------------------------------
    # CONSTRUCTOR
    # -----------------------------------

    def __init__(self):

        self.contadores = {}

    # -----------------------------------
    # FUNCION AGREGAR
    # -----------------------------------

    def agregar_contador(self):

        while True:

            nombre = input("\ningresa el nombre del contador: ")
            software = input("ingresa el software contable: ")
            clientes = input("ingresa la cartera de clientes: ")
            estados = input("ingresa los estados financieros: ")
            cedula = input("ingresa el numero de cedula: ")
            impuestos = input("ingresa los impuestos gestionados: ")
            bufete = input("ingresa el bufete: ")

            self.contadores[nombre] = {
                "software": software,
                "clientes": clientes,
                "estados": estados,
                "cedula": cedula,
                "impuestos": impuestos,
                "bufete": bufete
            }

            print("\ncontador agregado correctamente")

            respuesta = input("deseas agregar otro contador? si/no: ")

            if respuesta == "no":
                break

    # -----------------------------------
    # FUNCION LISTAR
    # -----------------------------------

    def listar_contadores(self):

        while True:

            if len(self.contadores) == 0:

                print("\nno hay contadores registrados")

            else:

                print("\n================ TABLA DE CONTADORES ================\n")

                print("{:<15} {:<18} {:<15} {:<20} {:<15} {:<15} {:<20}".format(
                    "NOMBRE",
                    "SOFTWARE",
                    "CLIENTES",
                    "ESTADOS",
                    "CEDULA",
                    "IMPUESTOS",
                    "BUFETE"
                ))

                print("=" * 130)

                for nombre, datos in self.contadores.items():

                    print("{:<15} {:<18} {:<15} {:<20} {:<15} {:<15} {:<20}".format(
                        nombre,
                        datos["software"],
                        datos["clientes"],
                        datos["estados"],
                        datos["cedula"],
                        datos["impuestos"],
                        datos["bufete"]
                    ))

                print("=" * 130)

            respuesta = input("\ndeseas volver a listar? si/no: ")

            if respuesta == "no":
                break

    # -----------------------------------
    # FUNCION ACTUALIZAR
    # -----------------------------------

    def actualizar_contador(self):

        while True:

            buscar = input("\ningresa el nombre del contador a actualizar: ")

            if buscar in self.contadores:

                nuevo_nombre = input("nuevo nombre: ")
                nuevo_software = input("nuevo software contable: ")
                nuevos_clientes = input("nueva cartera de clientes: ")
                nuevos_estados = input("nuevos estados financieros: ")
                nueva_cedula = input("nuevo numero de cedula: ")
                nuevos_impuestos = input("nuevos impuestos gestionados: ")
                nuevo_bufete = input("nuevo bufete: ")

                del self.contadores[buscar]

                self.contadores[nuevo_nombre] = {
                    "software": nuevo_software,
                    "clientes": nuevos_clientes,
                    "estados": nuevos_estados,
                    "cedula": nueva_cedula,
                    "impuestos": nuevos_impuestos,
                    "bufete": nuevo_bufete
                }

                print("\ncontador actualizado correctamente")

            else:

                print("\ncontador no encontrado")

            respuesta = input("\ndeseas actualizar otro contador? si/no: ")

            if respuesta == "no":
                break

    # -----------------------------------
    # FUNCION BORRAR
    # -----------------------------------

    def borrar_contador(self):

        while True:

            borrar = input("\ningresa el nombre del contador a borrar: ")

            if borrar in self.contadores:

                del self.contadores[borrar]

                print("\ncontador eliminado correctamente")

            else:

                print("\ncontador no encontrado")

            respuesta = input("\ndeseas borrar otro contador? si/no: ")

            if respuesta == "no":
                break


# ======================================================
# OBJETO
# ======================================================

sistema = contador()


# ======================================================
# MENU PRINCIPAL
# ======================================================

while True:

    print("\n================ MENU DE CONTADORES ================")
    print("1. agregar contador")
    print("2. listar contadores")
    print("3. actualizar contador")
    print("4. borrar contador")
    print("5. salir")

    opcion = int(input("\ningresa una opcion: "))

    if opcion == 1:

        sistema.agregar_contador()

    elif opcion == 2:

        sistema.listar_contadores()

    elif opcion == 3:

        sistema.actualizar_contador()

    elif opcion == 4:

        sistema.borrar_contador()

    elif opcion == 5:

        print("\nfin del programa")
        break

    else:

        print("\nopcion no valida")
```
